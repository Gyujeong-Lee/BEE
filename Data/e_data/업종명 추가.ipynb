## 표준산업분류 연계표 사용

```python
kospi200['업종대분류'] = pd.Series()
```

```python
industry_codes = pd.read_excel('./업종분류.xlsx')
```

```python
industry_codes.drop(columns=['Unnamed: 0', '대분류', '대분류.1', '중분류', '중분류.1', '소분류.1', '세분류.1', '세세분류', '연계', '대분류.2', '중분류.2', '중분류.3', '소분류.3', '세분류.3', '세세분류.1', '연계.1', '메인'], axis=1, inplace=True)
industry_codes
```

```python
industry_codes = industry_codes.fillna(0)
industry_codes
```

```python
industry_codes = industry_codes.astype({'2020년 귀속 업종코드':'int', '소분류':'int', '세분류':'int', '소분류.2':'int', '세분류.2':'int'})
industry_codes
```

```python
industry_codes.info()
```

```python
for i in range(1685):
    industry_codes.loc[i, '2020년 귀속 업종코드'] = str(format(industry_codes.loc[i, '2020년 귀속 업종코드'], '06'))
    industry_codes.loc[i, '소분류'] = str(format(industry_codes.loc[i, '소분류'], '03'))
    industry_codes.loc[i, '세분류'] = str(format(industry_codes.loc[i, '세분류'], '04'))
    industry_codes.loc[i, '소분류.2'] = str(format(industry_codes.loc[i, '소분류.2'], '03'))
    industry_codes.loc[i, '세분류.2'] = str(format(industry_codes.loc[i, '세분류.2'], '04'))
    
industry_codes = industry_codes.astype({'2020년 귀속 업종코드':'str', '소분류':'str', '세분류':'str', '소분류.2':'str', '세분류.2':'str'})
industry_codes
```

```python
# industry_codes['2020년 귀속 업종코드'] = industry_codes['2020년 귀속 업종코드'].astype(str)
# industry_codes['소분류'] = industry_codes['소분류'].astype(str)
# industry_codes['세분류'] = industry_codes['세분류'].astype(str)
# industry_codes['소분류.2'] = industry_codes['소분류.2'].astype(str)
# industry_codes['세분류.2'] = industry_codes['세분류.2'].astype(str)

industry_codes[['2020년 귀속 업종코드', '소분류', '세분류', '소분류.2', '세분류.2', '표준산업 분류']] = industry_codes[['2020년 귀속 업종코드', '소분류', '세분류', '소분류.2', '세분류.2', '표준산업 분류']].astype(str)

# industry_codes
```

```python
kospi200
```

```python
for i in range(200):
    code = kospi200.loc[i, '업종코드']
    l = len(code)
    if l == 3:
        if code in industry_codes['소분류'].values:
            kospi200.loc[i, '업종대분류'] = industry_codes[industry_codes['소분류'] == code]['대분류.3'].values[0]
        else:
            kospi200.loc[i, '업종대분류'] = industry_codes[industry_codes['소분류.2'] == code]['대분류.3'].values[0]
    elif l == 4:
        if code in industry_codes['세분류'].values:
            kospi200.loc[i, '업종대분류'] = industry_codes[industry_codes['세분류'] == code]['대분류.3'].values[0]
        else:
            kospi200.loc[i, '업종대분류'] = industry_codes[industry_codes['세분류.2'] == code]['대분류.3'].values[0]
    elif l == 5:
        kospi200.loc[i, '업종대분류'] = industry_codes[industry_codes['표준산업 분류'] == code]['대분류.3'].values[0]
        
kospi200
```

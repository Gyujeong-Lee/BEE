```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import requests

plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
stocks = pd.read_csv('./상장회사목록_20210907.csv', encoding='CP949')
environment = pd.read_excel('./2020년도 업체별 명세서 주요정보_21.6.25기준(정정 최종).xlsx')
```

```python
environment.drop(columns=['번호', '비고'], axis=1, inplace=True)
environment.dropna(inplace=True)
environment['종목명'] = pd.Series()
environment
```

```python
for i in range(1061):
    name = environment.loc[i, '관리업체']
    if '(주)' in name:
        name = name.replace('(주)', '')
    if '주식회사' in name:
        name = name.replace('주식회사', '')
    name = name.strip()

    if stocks[stocks['종목명']==name]['종목명'].values:
        environment.loc[i, '종목명'] = stocks[stocks['종목명']==name]['종목명'].values[0]

environment.info()
```

```python
data = pd.read_excel('./업종대분류추가.xlsx')
```

```python
data.head()
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
data['온실가스배출량'] = pd.Series()
data['에너지사용량'] = pd.Series()
data
```

```python
for i in range(200):
    data.loc[i, '종목코드'] = format(int(data.loc[i, '종목코드']), '06')
    data.loc[i, '고유번호'] = format(int(data.loc[i, '고유번호']), '08')
    
    company = environment[environment['종목명']==data.loc[i, '종목명']]
    
    if len(company.values):
        data.loc[i, '온실가스배출량'] = company['온실가스 배출량(tCO2)'].values[0]
        data.loc[i, '에너지사용량'] = company['에너지 사용량(TJ)'].values[0]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
codes = pd.read_excel('./data_3033_20210906.xlsx')
```

```python
for i in range(200):
    codes.loc[i, '종목코드'] = format(codes.loc[i, '종목코드'], '06')
            
codes
```

```python
# 고유번호 column 추가 (전부 0으로 초기화)
codes['고유번호'] = pd.Series()
codes
```

```python
import xml.etree.ElementTree as ET

tree = ET.parse('./CORPCODE.xml') # CORPCODE.xml을 파싱하여 tree에 저장
root = tree.getroot()

for i in range(200):
    code = codes.iloc[i]['종목코드']
    for company in root.findall('list'):
        stock_codes = company.find('stock_code').text
        if str(code) in stock_codes:
            codes.loc[i, '고유번호'] = str(company.find('corp_code').text)
            break
            
codes
```

```python
codes.to_excel('고유번호추가.xlsx')
```

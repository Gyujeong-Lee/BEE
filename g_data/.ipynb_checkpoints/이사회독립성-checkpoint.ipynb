```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import json
from pprint import pprint
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
data = pd.read_excel('./고유번호추가.xlsx')
```

```python
data
```

```python
unique_numbers = data['고유번호']
unique_numbers
```

```python
#dart open api 인자
auth_key="396b64cf35cf2138bbee974d47ca9d07baccc3e0" 
# 사외이사 비율
# 대표이사와 이사회 의장 일치 여부

board_res = []
representative_res = []
chairman_res = []
outside_director_ratio = []

for unique_number in unique_numbers:
    representative = [] # 대표이사
    chairman = '' # 이사회 의장
    outside_director = []
    
    reprt_code = 11011
    company_code = str(unique_number).zfill(8)
    url = 'https://opendart.fss.or.kr/api/exctvSttus.json?crtfc_key={}&corp_code={}&bsns_year=2020&reprt_code={}'.format(auth_key, company_code, reprt_code)
    res = requests.get(url).json()
    if res['status'] == '013':
        chk = False
        for i in range(1,4):
            url = 'https://opendart.fss.or.kr/api/exctvSttus.json?crtfc_key={}&corp_code={}&bsns_year=2020&reprt_code={}'.format(auth_key, company_code, reprt_code+i)
            res = requests.get(url).json()
            if res['status'] != '013':
                chk = True
                break
        if not chk:
            board_res.append(float('nan'))
            chairman_res.append(float('nan'))
            representative_res.append(float('nan'))
            outside_director_ratio.append(float('nan'))
            continue
    
    boards = res['list']
    pprint(boards)
    for board in boards:
        if ('이사회 의장' in board.get('chrg_job')) or ('이사회의장' in board.get('chrg_job')) or ('이사회\n의장' in board.get('chrg_job')):
            chairman = board.get('nm')
        if ('대표이사' in board.get('ofcps')) or ('대표이사' in board.get('chrg_job')) or ('CEO' in board.get('chrg_job')):
            representative.append(board.get('nm'))
        if ('사외이사' in board.get('ofcps')) or ('사외이사' in board.get('chrg_job')):
            outside_director.append(board.get('nm'))
    
    if chairman == '':
        board_res.append(float('nan'))
        representative_res.append(representative)
        chairman_res.append(chairman)
        outside_director_ratio.append(len(outside_director)/len(boards) * 100)
        continue
    else:
        if chairman in representative:
            board_res.append(True)
        else:
            board_res.append(False)
            
        representative_res.append(representative)
        chairman_res.append(chairman)
        outside_director_ratio.append(len(outside_director)/len(boards) * 100)
```

```python
for i in range(200):
    if type(representative_res[i]) == float:
        continue
    else:
        representative_res[i] = ','.join(representative_res[i])
```

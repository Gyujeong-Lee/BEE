```python
# (연결) 현금배당성향(%)
allocation_res = []
profit_res = []
for unique_number in unique_numbers:
    reprt_code = 11011
    allocation_url = f'https://opendart.fss.or.kr/api/alotMatter.json?crtfc_key=237b4436451d0a243043ac0f7d73993efb3dd834&corp_code={str(unique_number).zfill(8)}&bsns_year=2020&reprt_code={reprt_code}'
    res = requests.get(allocation_url).json()
    if res['status'] == '013':
        chk = False
        for i in range(1,4):
            allocation_url = f'https://opendart.fss.or.kr/api/alotMatter.json?crtfc_key=237b4436451d0a243043ac0f7d73993efb3dd834&corp_code={str(unique_number).zfill(8)}&bsns_year=2020&reprt_code={reprt_code+i}'
            print(str(unique_number).zfill(8))
            res = requests.get(allocation_url).json()
            if res['status'] != '013':
                chk = True
                break
        if not chk:
            allocation_res.append(float('nan'))
            profit_res.append(float('nan'))
            continue
            
    allocations = res['list']
    for allocation in allocations:
        print(allocation)
        if '(연결)당기순이익' in allocation.get('se') :
            profit_res.append(allocation.get('thstrm'))
        if '현금배당금총액' in allocation.get('se'):
            allocation_res.append(allocation.get('thstrm'))
            break
```

```python
# 직원보수
emp_res = [] # 직원 보수

for unique_number in unique_numbers:
    print(unique_number)
    reprt_code = 11011
    emp_compensation_url = f'https://opendart.fss.or.kr/api/empSttus.json?crtfc_key=237b4436451d0a243043ac0f7d73993efb3dd834&corp_code={str(unique_number).zfill(8)}&bsns_year=2020&reprt_code={reprt_code}'
    res = requests.get(emp_compensation_url).json()
    if res['status'] == '013':
        chk = False
        for i in range(1,4):
            emp_compensation_url = f'https://opendart.fss.or.kr/api/empSttus.json?crtfc_key=237b4436451d0a243043ac0f7d73993efb3dd834&corp_code={str(unique_number).zfill(8)}&bsns_year=2020&reprt_code={reprt_code+i}'
            res = requests.get(emp_compensation_url).json()
            if res['status'] != '013':
                chk = True
                break
        if not chk:
            emp_res.append((str(unique_number).zfill(8), '-', '-'))
            continue

    emp_compensations = res['list']
    temp = []
    for emp in emp_compensations:
        if emp['jan_salary_am'] != '-':
            temp.append((emp['sexdstn'], emp['jan_salary_am']))
    
    emp_res.append((str(unique_number).zfill(8), temp))
```

```python
emp_res
```

```python
emp_average = []
for e in emp_res:
    res = e[1]
    length = len(res)
    if not res or (length == 1 and e[1][0] == '-' or e[1][1] == '-'):
        emp_average.append(float('nan'))
        continue
    total = 0
    print(e[0])
    for i in range(length):
        total += int(res[i][1].replace(',',''))
        
    emp_average.append(total/length)
```

```python
emp_average
```

```python
data['직원보수'] = pd.Series(data=emp_average, dtype=str)

data
```
